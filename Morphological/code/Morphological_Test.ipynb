```python
from google.colab.patches import cv2_imshow
import os
import cv2 as cv
import matplotlib.pyplot as plt
import numpy as np

file_list = os.listdir('/content/drive/MyDrive/SIA/Morphological/DL/sampler/Dataset') #change point

test_png = []

#file name list up
for i in file_list:
  if '.png' in i:
    test_png.append(i)

#img opening and closing
for j in test_png:
  img_path = '/content/drive/MyDrive/SIA/Morphological/DL/sampler/Dataset/' + j #change point
  img = cv.imread(img_path, cv.IMREAD_COLOR)
  kernel = np.ones((5,5),np.uint8)
  erosion = cv.erode(img,kernel,iterations = 3) #optional change point
  dilation = cv.dilate(img,kernel,iterations = 3) #optional change point
  opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)
  closing = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)

  dilation_erosion = cv.erode(dilation, kernel, iterations = 3)
  erosion_dilation = cv.erode(erosion, kernel, iterations = 3)

  print(j)
  #save img
  #cv.imwrite('/content/drive/MyDrive/SIA/Morphological/DL/sampler/dilation > erosion/iteration3/'+j, dilation_erosion) #change point
  #cv.imwrite('/content/drive/MyDrive/SIA/Morphological/CDFL/dilation > erosion/iteration5'+j, erosion_dilation) #change point

  #show img
  plt.figure(figsize=(20,20)) #img size
  
  plt.subplot(131) #original img
  plt.axis('off')
  plt.imshow(img)

  plt.subplot(132) #opening img
  plt.axis('off')
  plt.imshow(dilation_erosion)

  plt.subplot(133) #closing img
  plt.axis('off')
  plt.imshow(closing)

  plt.show()
```
